
# Enfoque analítico basado en el tipo de pregunta

A la hora de elegir un enfoque analítico para un problema, el tipo de pregunta que se intenta responder influye enormemente en la metodología. He aquí cinco tipos habituales de preguntas y sus correspondientes enfoques analíticos:

## 1. Preguntas descriptivas: "¿Cuál es la situación actual?"
**Enfoque:** Análisis descriptivo  

**Pregunta:** "¿Cuál es el estado actual de nuestras ventas?"  

### Técnicas:
- **Agregación de datos:** Combinación de datos de varias fuentes en una vista unificada.
- **Minería de datos:** Extracción de información útil de grandes conjuntos de datos.
- **Visualización de datos:** Utilización de herramientas visuales para presentar los datos en un formato fácilmente comprensible.

### Ejemplos:
- Resumir datos de ventas
- Creación de cuadros de mando
- Generación de informes

---

## 2. Preguntas de diagnóstico: "¿Por qué ha ocurrido?"
**Enfoque:** Análisis de diagnóstico  

**Pregunta:** "¿Por qué disminuyeron nuestras ventas en el último trimestre?"  

### Técnicas:
- **Desglose:** Exploración de datos detallados para encontrar las causas subyacentes.
- **Descubrimiento de datos:** Identificación de patrones y relaciones en los datos.
- **Análisis de correlaciones:** Evaluación de la relación entre distintas variables.

### Ejemplos:
- Identificación de las causas raíz del descenso de las ventas
- Analizar las quejas de los clientes
- Comprender los puntos de fallo de un proceso

---

## 3. Preguntas predictivas: "¿Qué es probable que ocurra?"
**Enfoque:** Analítica predictiva  

**Pregunta:** "¿Cuál es nuestra previsión de ventas para el próximo año?"  

### Técnicas:
- **Análisis de regresión:** Predicción de resultados basada en las relaciones entre variables.
- **Predicción de series temporales:** Predicción de valores futuros a partir de tendencias pasadas.
- **Modelos de Aprendizaje automático:** Uso de algoritmos para predecir resultados futuros basándose en datos históricos.

### Ejemplos:
- Previsión de ventas
- Predicción de la pérdida de clientes
- Estimación de la demanda futura

---

## 4. Preguntas prescriptivas: "¿Qué debemos hacer?"
**Enfoque:** Análisis prescriptivo  

**Pregunta:** "¿Qué debemos hacer para aumentar el tráfico del sitio web?"  

### Técnicas:
- **Modelos de optimización:** Encontrar la mejor solución a partir de un conjunto de alternativas.
- **Simulación:** Modelización de escenarios para predecir resultados.
- **Análisis de decisiones:** Evaluación y comparación de diferentes decisiones.

### Ejemplos:
- Recomendación de niveles de inventario
- Optimización de campañas de marketing
- Determinación de estrategias de precios

---

## 5. Preguntas de clasificación: "¿A qué categoría pertenece esto?"
**Enfoque:** Clasificación (Aprendizaje supervisado)  

**Pregunta:** "¿A qué categoría pertenece este punto de datos?"  

### Técnicas:
- **Regresión logística:** Predicción de la probabilidad de un resultado categórico.
- **Árboles de decisiones:** Dividir los datos en ramas para clasificarlos.
- **Máquinas de vectores de asistencia:** Búsqueda de la mejor frontera para separar categorías.
- **Redes neuronales:** Utilización de nodos interconectados para clasificar datos.

### Ejemplos:
- Detección de correo basura
- Clasificación de imágenes
- Diagnóstico de enfermedades

---

**Comprender estos diferentes tipos de preguntas y los enfoques analíticos correspondientes puede ayudarle a liberar el verdadero potencial de sus datos.**
